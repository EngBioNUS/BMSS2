* to auto-rank the behaviors of the combinatorial gene circuit configurations

## Read .ini file & Add the model to database
Here is to read the .ini configuration file and store the model into the system database .db. More than one .ini files can be read and stored at once. Just define the filenames as [file1, file2, ....]

````python
filenames = ['Combinatorial_Bioconversion_Naringenin_comb.ini']
model, stored_model = mca.read_and_add_model_to_database(filenames)

print('\nThe model in the form of dictionary: \n', model[0]) # to view the first model, the index starts from 0

print('\nParameters: \n', model[0].get('parameters')) # to view the parameters' names available inside the model
````

## Generate model function in .py file to be used for Combinatorial Simulation
In order to run the model simulation, we need to first convert the model information (in the form of dictionary) into python function, so that the model in the form of ordinary differential equations (ODEs) can be solved using numerical solver. Set the input argument local = True to generate the .py file in the same directory, else local = False, the file will be generated in the global folder: BMSS/models/model_functions folder.

````python
modelstr, modelpath = mca.model_to_code(model[0], local=True) 

print('\nModel in string format:\n', modelstr)

print('\nModel path:\n', modelpath)
````

## Get the model ODE function from the generated .py file. 

````python
modelfun = mca.get_model_function(modelpath)

print(modelfun)
````

## Define the initial conditions for states
Define in the dictionary format for ease of tuning, follow the order of states given in the .ini 

````python
y0dict = {'Inde':0, 'Indi': 0, 'mCHS':0, 'OsCHS':0, \
          'mMCS':0, 'MCS':1E-06, \
          'mPAL':0, 'PAL':1E-06, \
          'mE4CL':0, 'E4CL': 1E-06, \
          'Tyr':0, 'CouA':0, 'CouCoA':0, 'MalA':0, 'MalCoA':0, 'Nar':0}

# Get the y0 in list for ODE solving
y0 = list(y0dict.values())
````

## Define parameters' values and modify if necessary
Define in dictionary format for ease of manipulation, follow the order of parameters provided in the .ini file. There is also function where users can use to modify parameters by multiplying with certain value. An example is shown below. 

````python
paramsdict = {'Vm': 6.1743e-05, 'ntrans': 0.9416,\
              'Ktrans': 0.0448, 'n_ind': 5.4162, 'K_ind': 2.0583e-05,\
              'syn_mRNA': 8.6567e-08, 'syn_Pep': 0.01931 *0.9504,\
              'deg_Pep1': 0.0010, 'deg_mRNA': 0.1386, 'deg_Pep': 0.007397,\
              'syn_mMCS': 2.2953e-07, 'syn_pMCS': 0.01931 *0.9218,\
              'syn_mPAL': 2.2953e-07, 'syn_pPAL': 0.01931 *0.8684,\
              'syn_mE4CL': 2.2953e-07, 'syn_pE4CL': 0.01931 *0.9119,\
              'kcatTyr': 61.2 , 'KmTyr': 0.195e-3, 'kcatCouA': 16.92,\
              'KmCouA': 0.246e-3, 'kcatMalA': 150, 'KmMalA': 529.4e-6,\
              'kcatCouMalCoA': 0.0517, 'KmMalCoA': 47.42e-6,\
              'KmCouCoA': 45.44e-6}

plasmidcopy_fold = 5

# specify the parameter to be modified
modified_keylist = ['syn_mRNA', 'syn_mMCS', 'syn_mPAL', 'syn_mE4CL']

# multiple those parameters with the plasmidcopy_fold, assuming plasmid copy affects the mRNA synthesis rate
new_paramsdict = cm.multiply_dictkey_byvalue(paramsdict, modified_keylist,\
                                             plasmidcopy_fold)

print(new_paramsdict)

# Get the parameters' values in list for ODE solving
params = list(new_paramsdict.values())
print("\nParameters' values:\n", params)
````

## Define the time, state's event or parameter's event
The state and/or param event allow users to simulate the conditions when adding inducers (if represented as a state or a parameter) or substrates at certain time points (i.e. after growth phase etc.) 

````python
# cm.set_time(tstart, tend)
Time = cm.set_time(0, (18+24)*60) # in min

print('Time points: ', Time, 'length: ', Time.shape)

# assume the first event is adding 0.02 mM IPTG inducer (represented by index 0->Inde) at the beginning of the
# experiment (t=1), and the second event is adding the two substrates (0.25 mM Tyrosine-index -6 if counted backward,
# or index 13 if in counted in forward direction; 1mM Malonic Acid-index -3 or index 13) after 18 hours (18*60 mins).
# it doesn't matter if we count from forward or backward direction.
# (forward: start from 0; backward: from -1).
stateevent = {'Time':[1, 18*60], 'State':[{0: 0.02e-3}, {-6: 0.25e-3, -3: 1e-3}]}

# the same defining approach can be used to add paramevent, in this case, we assume there is no paramevent.
#paramevent = {'Time':[18*60, 24*60], 'Param':[{6: 1e-08}, {11: 1.2953e-07, 13: 0.2953e-07}]}
paramevent = {}
````

## Solve the model with Single Combination
This function allows users to simulate the ODE model with a single combination; can be used for testing purposes. The option of stateevent or/and paramevent is/are available to add inducers or substrates as shown above. __*Note:*__ Backslash \ at the end of a line is to extend the current logical line over to the next physical line. 

````python
comb = [1, 0.5, 0.3, 1, 0.8, 0.4, 0.9]
print('Combinations: ', comb) # Normally it represents the relative strength, but can be used to insert other values as well.

y = cm.solveODE_event(modelfun, y0, Time, args = (params, comb),\
                      stateevent = stateevent, paramevent = paramevent)

# Plot the result figures
# pf.plotfig(fignum, x, y, xylabels, legend, ...), this function also allows plotting of errorbar.
# Use same figure number to overlap multiple lines in the same figure (only applicable in python script, not jupyter notebook). 
# Or you can use 2D array for y as well.

# column stack the 1D array into 2D array for non-sequential column data plotting in a same figure 
AllEnzymes = np.column_stack((y[:,3], y[:,5], y[:,7], y[:,9]))

pf.plotfig(10, Time, y[:,-1], ['Time (min)', 'Concentration (Molar)'], ['Nar'])
pf.plotfig(11, Time, AllEnzymes, ['Time (min)', 'Concentration (Molar)'], ['OsCHS', 'MCS', 'PAL', '4CL'])
````

## Solve the model with All generated Combinations

### Define the relative strengths for promoters or/and RBSs

````python
# Define RBS Relative Strengths in dict
RBSdict = {'rbsD': 1.0, 'rbs32': 0.1855}

# Define Promoter Relative Strengths wrt J23102 in dict
Promoterdict = {'J23101': 0.909, 'J23106': 0.3636}

# Plot the relative strength partdict as heatmap and export the figure into outputpath. 
cm.plot_relative_strength({'PromoterOpt1':Promoterdict, 'RBSOpt1':RBSdict}, outputpath)
````

### Set the options for the different parts and Get all the generated combinations

````python
Optiondict = {'pMCS': Promoterdict, 'rbsMCS': RBSdict,
              'pPAL': Promoterdict, 'rbsPAL': RBSdict, 
              'p4CL': Promoterdict, 'rbs4CL': RBSdict,
              'rbsOsCHS': RBSdict}
dfcombs_str, dfcombs_val = cm.get_combinations(Optiondict)

# show the first five combinations of the dataframe in string (i.e. part name)
print(dfcombs_str.head(), '\n')

# show the first five combinations of the dataframe in value (i.e. relative strength value)
print(dfcombs_val.head(), '\n')

# show the info of the dataframe (total entry, columns, datatype for each columns)
print(dfcombs_val.info())
````

### Solve all combinations and return the time-response data for all the states by construct index

````python
states = list(y0dict.keys())  # this is to get the states and use as keys to store the time-response simulation results in the solution dictionary.
soldict = cm.solve_combinations(dfcombs_val, modelfun, states, y0,\
                                Time, params, stateevent = stateevent,\
                                paramevent = {})

print(soldict) # a nested dictionary with first-layer key as construct index, and second-layer key as state with their time-response data in an 1D array. 
````

### Get the time-response data for all indexes for a list of selected components/states

````python
compdict = cm.get_component(soldict, ['CouA', 'Nar'])
print(compdict) 
print('\nThe final values of Naringenin for all constructs:\n', compdict['Nar'][-1,:]) # get the end-point values of Naringenin for all constructs' indexes 
````

The reason for using compdict as input argument instead of soldict is actually to expedite the ranking process. 

````python
index, comparray, dfcombs_val_Output, dfcombs_str_Output = \
cm.rank_combinations(dfcombs_val, dfcombs_str, compdict, 'Nar', order = 'descending')

print(dfcombs_val_Output.head(), '\n')
print(dfcombs_str_Output.head(), '\n')
````

## Run sensitivity analysis and auto-plot the results in heatmap
__*num_vars*__: the number of variables to analyze, in this case, there are seven combinatorial variables.

__*names*__: the names of the combinatorial variables to be shown on the output heatmap.

__*bounds*__: are set to be within [0, 1] for all variables.

__*method*__: There are four methods available now: 'sobol', 'fast', 'RBD-FAST', 'delta'.

__*sampling*__: number of samples to withdraw from the bounds for running the sensitivity analysis. The default value is 1000. We set 100 here to shorten the time taken for running, but the accuracy will be lower.

* For 'sobol' method, there will be two output figures: one is the first-order sensitivities dictating how each variable affects the output variation, the other is the second-order sensitivities showing how the interactions of variables affect the output variations. 

````python
xtlabels = list(Optiondict.keys())
problem = {
        'num_vars': 7,
        'names': xtlabels,
        'bounds': [[0, 1]]*7
        }

Si = cm.run_sensitivity_analysis(problem, modelfun, y0, Time, params,\
                         stateevent = stateevent, paramevent = {},\
                         method = 'sobol', sampling = 100)
print(Si)
````

## Auxiliary Functions:
### Plot the gene circuit visual diagram and export the figure
__*partstr*__: a string containing the individual type of part, followed by color and name, separated by a space. Default color of black is used if not specified.
* The short forms of the different types of parts: p-> Promoter, r-> RBS, c-> CDS, t-> Terminator, s-> Spacer, o-> Origin

__*regulations*__: a string containing the 'from_part' to the 'to_part' connected by an arrow. Type of interaction is specified after the 'to_part' followed by the color. 
* The available interactions include 'Activation', 'Repression', 'Connection', 'Derepression', and 'Activation2'.
* The 'Repression' must be used before specifying 'Derepression', and 'Activation' must be used before specifying 'Activation2'. Example: 'c0->p0.Repression p0->p0.Derepression'    

````python
partstr = 'p.pLac r.rbsD c.blue.OsCHS t p r c.purple.MCS t p r c.orange.PAL t p r c.red.4CL t o.pRFS101'
regulations = 'p0->p0.Activation'
circuitfigpath = os.path.join(outputpath, 'Circuitfigure.tiff')
fig = pc.plot_circuit_config(partstr, regulations, savefig = circuitfigpath)
fig.show()
````
